```python
import pandas as pd
```

```python
data = pd.read_csv('/content/Airbnb_Open_Data.csv')
data.head()
```

```python
data.columns
```

```python
# Menghapus kolom yang tidak diperlukan

data_cleaned = data.drop(columns={'id','host id','country','service fee','license','house_rules'})
data_cleaned.head()
```

## **Mengatasi nilai yang hilang**

```python
# Menghapus baris dengan nilai yang hilang
# dropna(thresh=...) menghapus baris yang memiliki nilai yang hilang lebih dari jumlah kolom tertentu
# data_cleaned.shape[1]-2 baris yang memiliki lebih dari 2 nilai akan dihapus

data_cleaned = data_cleaned.dropna(thresh=data_cleaned.shape[1]-2)
```

```python
# Mengisi nilai yang hilang pada kolom 'reviews per month'
# Menggunakan nilai median karena lebih tahan terhadap outliners dibanding rata-rata

data_cleaned['reviews per month'].fillna(data_cleaned['reviews per month'].median(), inplace=True)
data_cleaned
```

```python
# Mengisi niai yang hilang pada kolom 'last review'

data_cleaned['last review'].fillna('No Review',inplace=True)
data_cleaned
```

## **Mengkonversi tipe data**

```python
data_cleaned.dtypes
```

```python
# Mengkonversi kolom tanggal
data_cleaned['last review']=pd.to_datetime(data_cleaned['last review'],errors = 'coerce')
data_cleaned['Construction year']=pd.to_datetime(data_cleaned['Construction year'],errors = 'coerce')
data_cleaned['Construction year'] = data_cleaned['Construction year'].dt.year
```

```python
data_cleaned.dtypes
```

```python
# Mengisi nilai yang hilang pada kolom numeric
data_cleaned['calculated host listings count'].fillna(0,inplace=True)
data_cleaned['availability 365'].fillna(0,inplace=True)
data_cleaned['minimum nights'].fillna(0,inplace=True)
```

```python
data_cleaned['calculated host listings count'] = data_cleaned['calculated host listings count'].astype(int)
data_cleaned['availability 365'] = data_cleaned['availability 365'].astype(int)
data_cleaned['minimum nights'] = data_cleaned['minimum nights'].astype(int)
data_cleaned.dtypes
```

```python
data_cleaned.head(30)
```

## **Menormalisasi Nilai**

```python
# Mengubah huruf menjadi title case
data_cleaned['neighbourhood group']= data_cleaned['neighbourhood group'].str.title()
data_cleaned['neighbourhood']= data_cleaned['neighbourhood'].str.title()

# Menghapus spasi berlebih
data_cleaned['neighbourhood group']= data_cleaned['neighbourhood group'].str.strip()
data_cleaned['neighbourhood']= data_cleaned['neighbourhood'].str.strip()

data_cleaned['neighbourhood group'].fillna('',inplace=True)
data_cleaned['neighbourhood'].fillna('',inplace=True)
data_cleaned.head(20)
```

```python
#data_cleaned = data_cleaned.drop(columns={'country code'})
data_cleaned['host_identity_verified'].fillna('unconfirmed',inplace=True)
data_cleaned['instant_bookable'].fillna('unconfirmed',inplace=True)
data_cleaned['instant_bookable']=data_cleaned['instant_bookable'].replace({'False':'No','True':'Yes'})
data_cleaned['review rate number'].fillna(data_cleaned['review rate number'].median(),inplace=True)
data_cleaned.head(25)
```

```python
data_cleaned['instant_bookable'] = data_cleaned['instant_bookable'].astype(str)
data_cleaned['instant_bookable']=data_cleaned['instant_bookable'].replace({'False':'No','True':'Yes'})
data_cleaned.head()
```

```python
# Memeriksa Duplikasi

data_cleaned = data_cleaned.drop_duplicates()
data_cleaned.head()
```

```python
# mengubah price menjadi integer
data_cleaned['price']=data_cleaned['price'].str.replace('$','')
data_cleaned['price'] = pd.to_numeric(data_cleaned['price'], errors='coerce').fillna(0).astype(int)
data_cleaned.dtypes
```

```python
data_cleaned
```

```python
# Menyimpan data yang telah dibersihkan ke file baru
cleaned_file_path = 'Airbnb_Cleaned_Construction_Year.csv'
data_cleaned.to_csv(cleaned_file_path, index=False)
```
